# Inputs

`VarKeep_Spatial`: Variance to keep when applying PCA to model variable $m$.

`Index`: Index of specified responses from 3DSL. For example, for Oilrate `Index.Oilrate` is set to 4 because the values are located at 4th column.


````matlab
VarKeep_Spatial = 95;

Index.Oilrate=4;
Index.Watercut=8;
Index.Time =2; 
MaxTimeHistory = 7500;
ReferenceNumber = 106;
FigSize_outputs = [319 215 1245 642];


WellNamesAll = {'P1','P2','P3','P4','P5','PNEW2'};
WellNamesData = {'P1','P2','P3','P4','P5'};
WellNamesPrediction = {'PNEW2'};
````
